```python
def visInterval(imgFolder, antFolder, prefix, pre=5, nex=7):
    path     = "/home/jovyan/data-vol-2/recycling/backboneData"
    imgPathL = sorted(glob.glob(f"{path}/{imgFolder}/*.jpg"))
    antPathL = sorted(glob.glob(f"{path}/{antFolder}/*.txt"))
    i = imgPathL.index(f"{path}/{imgFolder}/{prefix}.jpg")
    j = antPathL.index(f"{path}/{antFolder}/{prefix}.txt")
    assert i==j, (i,j)
    L = []
    for k,(imgPath,antPath) in enumerate(zip(imgPathL[i-pre:i+nex+1], antPathL[i-pre:i+nex+1])):
        print(k)
        vis.show(imgPath, antPath)
        L.append( os.path.basename(imgPath) )
    return L
    
visInterval("2022-06-30", "2022-06-30_label_tetra_v2", "20220630_140330_855")
```
